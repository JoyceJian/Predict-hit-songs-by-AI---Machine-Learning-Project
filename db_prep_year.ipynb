## Billboard Ranking Extract & Yearly DataFrame Set Up

```python
import billboard
from time import sleep
```

### 1990 Hot 100 Chart

```python
i = 1
charts1990 = []
while i <= 12:

  if i < 10:
    date1990 = "1990-0" + str(i) +"-01"
  else:
    date1990 = "1990-" + str(i) +"-01"

  charts = billboard.ChartData('hot-100', date=date1990, fetch=True, timeout=25)
  # print(charts)
  charts1990.append(charts)
  print(f"{date1990} chart appended.")

  i += 1
  sleep(180)
```

```python
charts1990
```

### Get Song Titles for 1990's chart

```python
# Year 1990
song_titles_1990 = []

for chart in charts1990:
    count = 0
    while count < len(chart):
        song_titles_1990.append(chart[count].title)
        count += 1
```

```python
len(song_titles_1990)
```

```python
# Spotify App Credentials 1990
CLIENT_ID = "aaf3f0bbaff74d50a1b8a244e4364bdf"
CLIENT_SECRET = "832f4869770d40ab9d503d1eeaaba8ce"

credentials = oauth2.SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET)

token = credentials.get_access_token()
spotify = spotipy.Spotify(auth=token)
```

```python
output_1990 = []
for title in song_titles_1990:
  output_1990.append(spotify.search(title, type="track", market=market, limit=1))
```

```python
ids_1990 = []
for track in output_1990:
    # remove tracks that doesn't have corresponding id on Spotify
    if len(track['tracks']['items']) >= 1: 
        ids_1990.append(track['tracks']['items'][0]['id'])
```

```python
len(ids_1990)
```

```python
i = 0
features_1990 = []
while i < len(ids_1990):
    if i+100 < len(ids_1990):
        features_1990.append(spotify.audio_features(ids_1990[i:i+100]))
    else:
        features_1990.append(spotify.audio_features(ids_1990[i:]))
#     print(i)
    i += 100
```

```python
features_1990_cleaned = []

for feature in features_1990:
#    print(len(feature))
     for lst in feature:
         if lst != None:
             features_1990_cleaned.append(lst)
```

```python
df_1990 = pd.DataFrame(features_1990_cleaned)
```

```python
df_1990.head()
```

```python
len(df_1990)
```

```python
df_1990.to_csv("Resources/1990data_final.csv")
```

```python
df_1990_cleaned = df_1990.drop_duplicates(subset="id", keep='first', inplace=False)
```

```python
df_1990_cleaned.to_csv("Resources/1990data_cleaned.csv")
```

## List Distinct Song Titles for Each Year

```python
def distinct(list): 
    
    distinct_items = [] 
    for item in list: 
        if item not in distinct_items: 
            distinct_items.append(item) 
    
    return distinct_items
```
